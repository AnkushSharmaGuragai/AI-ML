```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (int): The type of conversion selected by the user.
                           1 = meters to feet
                           2 = feet to meters
                           3 = kilograms to pounds
                           4 = pounds to kilograms
                           5 = liters to gallons
                           6 = gallons to liters
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:
        return value * 3.28084  # meters to feet
    elif conversion_type == 2:
        return value / 3.28084  # feet to meters
    elif conversion_type == 3:
        return value * 2.20462  # kg to pounds
    elif conversion_type == 4:
        return value / 2.20462  # pounds to kg
    elif conversion_type == 5:
        return value * 0.264172  # liters to gallons
    elif conversion_type == 6:
        return value / 0.264172  # gallons to liters
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    try:
        choice = int(input("Choose conversion type (1-6): "))
        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is None:
            print("Unsupported conversion type.")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Error: Please enter a valid number.")


if __name__ == "__main__":
    main()
```

```python
def calculate_sum(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
    numbers (list of float): List of numbers to sum.

    Returns:
    float: Sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculate the average of a list of numbers.

    Parameters:
    numbers (list of float): List of numbers to average.

    Returns:
    float: Average of the numbers.
    """
    return sum(numbers) / len(numbers)


def calculate_max(numbers):
    """
    Find the maximum value in a list of numbers.

    Parameters:
    numbers (list of float): List of numbers.

    Returns:
    float: Maximum number in the list.
    """
    return max(numbers)


def calculate_min(numbers):
    """
    Find the minimum value in a list of numbers.

    Parameters:
    numbers (list of float): List of numbers.

    Returns:
    float: Minimum number in the list.
    """
    return min(numbers)


def main():
    print("Mathematical Operations on a List of Numbers")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Choose an operation (1-4): "))
        numbers_input = input("Enter a list of numbers separated by spaces: ").strip()

        if not numbers_input:
            raise ValueError("Empty input not allowed.")

        # Convert input string to list of floats
        numbers = [float(num) for num in numbers_input.split()]

        if choice == 1:
            result = calculate_sum(numbers)
        elif choice == 2:
            result = calculate_average(numbers)
        elif choice == 3:
            result = calculate_max(numbers)
        elif choice == 4:
            result = calculate_min(numbers)
        else:
            print("Invalid operation selected.")
            return

        print("Result:", result)

    except ValueError as ve:
        print("Error:", ve)


if __name__ == "__main__":
    main()
```

```python
def extract_every_other(lst):
    """
    Extract every other element from a list, starting from the first element.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list containing every other element from the original list.
    """
    return lst[::2]  # Take every 2nd element starting from index 0


# Example usage:
original_list = [1, 2, 3, 4, 5, 6]
result = extract_every_other(original_list)
print("Original list:", original_list)
print("Every other element:", result)
```

```python
def get_sublist(lst, start, end):
    """
    Return a sublist from a given list, from start index to end index (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): Starting index (inclusive).
    end (int): Ending index (inclusive).

    Returns:
    list: A sublist from start to end index.
    """
    # Ensure end+1 because Python slicing is exclusive at the end
    return lst[start:end+1]


# Example usage:
original_list = [1, 2, 3, 4, 5, 6]
start_index = 2
end_index = 4
result = get_sublist(original_list, start_index, end_index)
print("Original list:", original_list)
print(f"Sublist from index {start_index} to {end_index}:", result)
```

```python
def reverse_list(lst):
    """
    Reverse a list using slicing.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list with elements in reverse order.
    """
    return lst[::-1]  # Slicing with step -1 reverses the list


# Example usage:
original_list = [1, 2, 3, 4, 5]
reversed_list = reverse_list(original_list)

print("Original list:", original_list)
print("Reversed list:", reversed_list)
```

```python
def remove_first_last(lst):
    """
    Remove the first and last elements from a list using slicing.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A sublist without the first and last elements.
    """
    return lst[1:-1]  # Slice from index 1 to one before the last element


# Example usage:
original_list = [1, 2, 3, 4, 5]
result = remove_first_last(original_list)

print("Original list:", original_list)
print("List without first and last elements:", result)
```

```python
def get_first_n(lst, n):
    """
    Extract the first n elements from a list using slicing.

    Parameters:
    lst (list): The original list.
    n (int): Number of elements to extract from the beginning.

    Returns:
    list: A sublist containing the first n elements.
    """
    return lst[:n]  # Slice from start up to index n (exclusive)


# Example usage:
original_list = [1, 2, 3, 4, 5]
n = 3
result = get_first_n(original_list, n)

print("Original list:", original_list)
print(f"First {n} elements:", result)
```

```python
def get_last_n(lst, n):
    """
    Extract the last n elements from a list using slicing.

    Parameters:
    lst (list): The original list.
    n (int): Number of elements to extract from the end.

    Returns:
    list: A sublist containing the last n elements.
    """
    return lst[-n:]  # Slice from -n to the end of the list


# Example usage:
original_list = [1, 2, 3, 4, 5]
n = 2
result = get_last_n(original_list, n)

print("Original list:", original_list)
print(f"Last {n} elements:", result)
```

```python
def reverse_skip(lst):
    """
    Extract every second element from a list in reverse order,
    starting from the second-to-last element.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list containing every second element from the end, moving backward.
    """
    return lst[-2::-2]  # Start at second-to-last (-2), move backwards with step -2


# Example usage:
original_list = [1, 2, 3, 4, 5, 6]
result = reverse_skip(original_list)

print("Original list:", original_list)
print("Elements in reverse order, skipping one:", result)
```

```python
def flatten(lst):
    """
    Flatten a nested list into a single list.

    Parameters:
    lst (list of lists): A nested list where each element is a list.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []
    for sublist in lst:
        flat_list.extend(sublist)  # Add all elements of sublist to flat_list
    return flat_list


# Example usage:
nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)

print("Nested list:", nested_list)
print("Flattened list:", result)
```

```python
def access_nested_element(lst, indices):
    """
    Access a specific element from a nested list using a list of indices.

    Parameters:
    lst (list of lists): The nested list.
    indices (list of int): List of indices specifying the path to the element.

    Returns:
    The element at the specified position in the nested list.
    """
    element = lst
    for index in indices:
        element = element[index]  # Drill down to the next level
    return element


# Example usage:
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)

print("Nested list:", nested_list)
print(f"Element at indices {indices}:", result)
```

```python
def sum_nested(lst):
    """
    Calculate the sum of all elements in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    int/float: The sum of all numeric elements in the nested list.
    """
    total = 0
    for element in lst:
        if isinstance(element, list):
            total += sum_nested(element)  # Recursively sum sublists
        else:
            total += element  # Add numbers directly
    return total


# Example usage:
nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

```python
def remove_element(lst, elem):
    """
    Remove all occurrences of a specific element from a nested list.

    Parameters:
    lst (list of lists): The nested list.
    elem: The element to remove.

    Returns:
    list: A new nested list with the element removed from all sublists.
    """
    new_list = []
    for sublist in lst:
        if isinstance(sublist, list):
            # Remove the element from sublist using list comprehension
            filtered = [x for x in sublist if x != elem]
            new_list.append(filtered)
        else:
            # If the element itself is not a list, just keep it if it's not the element to remove
            if sublist != elem:
                new_list.append(sublist)
    return new_list


# Example usage:
nested_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2
result = remove_element(nested_list, element_to_remove)

print("Original nested list:", nested_list)
print(f"Nested list after removing {element_to_remove}:", result)
```

```python
def find_element_max(lst):
    """
    Find the maximum element in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    int/float: The maximum element in the nested list.
    """
    max_val = float('-inf')  # Start with the smallest possible value
    for element in lst:
        if isinstance(element, list):
            # Recursively find the maximum in the sublist
            sub_max = find_element_max(element)
            if sub_max > max_val:
                max_val = sub_max
        else:
            if element > max_val:
                max_val = element
    return max_val


# Example usage:
nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_element_max(nested_list)

print("Nested list:", nested_list)
print("Maximum element:", result)
```

```python
def count_occurrences(lst, elem):
    """
    Count how many times a specific element appears in a nested list.

    Parameters:
    lst (list): A nested list containing numbers or other lists.
    elem: The element to count.

    Returns:
    int: Number of times elem appears in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            # Recursively count occurrences in sublist
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


# Example usage:
nested_list = [[1, 2], [2, 3], [2, 4]]
element_to_count = 2
result = count_occurrences(nested_list, element_to_count)

print("Nested list:", nested_list)
print(f"Occurrences of {element_to_count}:", result)
```

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list (any depth) into a single list.

    Parameters:
    lst (list): A nested list of arbitrary depth.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []
    for element in lst:
        if isinstance(element, list):
            # Recursively flatten sublists
            flat_list.extend(deep_flatten(element))
        else:
            flat_list.append(element)
    return flat_list


# Example usage:
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print("Deeply nested list:", nested_list)
print("Flattened list:", result)
```

```python
def average_nested(lst):
    """
    Calculate the average of all elements in a nested list.

    Parameters:
    lst (list): A nested list containing numbers or other lists.

    Returns:
    float: The average of all numeric elements in the nested list.
    """
    def sum_count(sublist):
        """Helper function to recursively sum elements and count them."""
        total = 0
        count = 0
        for element in sublist:
            if isinstance(element, list):
                sub_total, sub_count = sum_count(element)
                total += sub_total
                count += sub_count
            else:
                total += element
                count += 1
        return total, count

    total_sum, total_count = sum_count(lst)
    if total_count == 0:
        return 0  # Avoid division by zero for empty list
    return total_sum / total_count


# Example usage:
nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print("Nested list:", nested_list)
print("Average of all elements:", result)
```

```python
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("1. A * A^-1 =\n", identity_check)
print("Check if identity (approx):", np.allclose(identity_check, np.eye(2)))

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB =\n", AB)
print("BA =\n", BA)
print("Check AB != BA:", not np.allclose(AB, BA))

# 3. Prove (AB)^T = B^T * A^T
AB_T = (np.dot(A, B)).T
B_T_A_T = np.dot(B.T, A.T)
print("\n3. (AB)^T =\n", AB_T)
print("B^T * A^T =\n", B_T_A_T)
print("Check equality:", np.allclose(AB_T, B_T_A_T))

# ------------------------------------------------------------
# Solve the system of linear equations using inverse method
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent in matrix form AX = B
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Solve using inverse method
A_eq_inv = np.linalg.inv(A_eq)
X = np.dot(A_eq_inv, B_eq)
print("\nSolution of the system [x, y, z]:", X)

# Optional: Solve using np.linalg.solve (more stable)
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve:", X_solve)
```
